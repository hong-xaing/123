```python
import requests
from bs4 import BeautifulSoup
import time
from openpyxl import Workbook

# 创建一个新的工作簿
wb = Workbook()
# 选择活动的工作表
ws = wb.active
# 添加标题行
ws.append(["標題", "連結"])

page = 1
while True:
    print("現在抓到第" + str(page) + "頁")
    url = "https://gimy.app/cat/4--------" + str(page) + "---.html"
    res = requests.get(url)
    soup = BeautifulSoup(res.text, "lxml")

    # 找到包含链接的元素
    html2_list = soup.find_all("div", class_="box-video-list")[0].find_all("div", class_="title")

    # 遍历每个包含链接的元素并提取链接
    for href in html2_list:
        a_tag = href.find("h5").find("a")
        if a_tag:
            link = a_tag["href"]
            title_text = a_tag["title"]
            # 将数据添加到工作表
            ws.append([title_text, "https://gimy.app/"+link])

    page += 1
    time.sleep(2)

# 保存工作簿到文件
    wb.save("gimy抓取的数据.xlsx")
```

```python
import requests
from bs4 import BeautifulSoup

# 初始化页面计数器
page = 1

while True:
    # 构建当前页面的URL
    url = f"https://gimy.app/cat/4--------{page}---.html"
    
    # 发送HTTP请求
    res = requests.get(url)
    
    # 检查是否成功获取页面
    if res.status_code != 200:
        break  # 如果无法获取页面，退出循环
    
    soup = BeautifulSoup(res.text, "lxml")
    print(f"现在抓到第 {page} 页")
    
    # 找到包含链接的元素
    html2_list = soup.find_all("div", class_="layout-box clearfix")
    
    # 检查是否有足够的匹配项
    if len(html2_list) >= 2:
        # 遍历匹配项，提取第二个匹配项
        for i in range(1, len(html2_list)):
            match = html2_list[i]
            a_tag = match.find("h5").find("a")
            if a_tag:
                link = a_tag["href"]
                title = a_tag["title"]
                print(f"标题: {title}, 链接: {link}")
    
    page += 1

    
```

```python
import requests
from bs4 import BeautifulSoup
import time

page = 1
while True:
    print("現在抓到第"+str(page)+"頁")
    url = "https://gimy.app/cat/4--------"+str(page)+"---.html"
    res = requests.get(url)
    soup = BeautifulSoup(res.text,"lxml")
    
    # 使用 CSS 选择器来找到所需元素
    titles = soup.select(".box-video-list .title h5 a")
    
    for title in titles:
        link = title["href"]
        title_text = title["title"]
        print(f"標題: {title_text}, 連結: {link}")
        
    page += 1
    time.sleep(2)
```
