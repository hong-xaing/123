```python
import cv2
import numpy as np

kings = ['Andy Lau','Aaron Kwok',"Leon Lai","Jacky Cheung"]
kings_en=["L","Q","M","Z"]
img_train=[]
labels=[]

king_num= 0
for king in kings_en:
    
    num=0
    while num <10:
        num+=1
        img_train.append(cv2.imread(king+str(num)+".JPG",cv2.IMREAD_GRAYSCALE))
        labels.append(king_num)
    king_num+=1
# print(labels)

recon = cv2.face.LBPHFaceRecognizer_create()
recon.train(img_train,np.array(labels))

img = cv2.imread("4kings.jpg")
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
faceCascade = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")

faces = faceCascade.detectMultiScale(
    gray,
    scaleFactor=1.12,
    minNeighbors=3,
    minSize=(50,50)
)
print(faces)
print("發現",len(faces),"個人臉！！")
cut_list = []
face_num = 0

for (x, y, w, h) in faces:
    face_num += 1

    center_x = x + w // 2
    center_y = y + h // 2

    crop_width = crop_height = max(w, h) // 2
    crop_x = max(center_x - crop_width // 2, 0)
    crop_y = max(center_y - crop_height // 2, 0)

    crop_img = img[crop_y:crop_y + crop_height, crop_x:crop_x + crop_width]

    cv2.imwrite("cut" + str(face_num) + ".jpg", crop_img)
    print("已儲存", "cut" + str(face_num) + ".jpg")
    cut_list.append("cut" + str(face_num) + ".jpg")

# print(cut_list)
cut_who = []
for cut in cut_list:
    predict_img = cv2.imread(cut,cv2.IMREAD_GRAYSCALE)
    
    label,confidence = recon.predict(predict_img)
    cut_who.append(kings[label])
    
    print("此人是",kings[label],
          "confidence=",confidence)
who_num = 0
for (x,y,w,h) in faces :
    cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),2)
    cv2.putText(img, cut_who[who_num], (x, y - 10), cv2.FONT_HERSHEY_TRIPLEX, 1, (0, 255, 0), 1)
    who_num+=1

cv2.imshow("findface",img)
cv2.waitKey()
```

```python
import cv2
import numpy as np

kings = ['Andy Lau', 'Aaron Kwok', 'Leon Lai', 'Jacky Cheung']
kings_en = ["L", "Q", "M", "Z"]
img_train = []
labels = []

king_num = 0
for king in kings_en:
    num = 0
    while num < 10:
        num += 1
        img_train.append(cv2.imread(king + str(num) + ".JPG", cv2.IMREAD_GRAYSCALE))
        labels.append(king_num)
    king_num += 1

recon = cv2.face.LBPHFaceRecognizer_create()
for king_en in kings_en:
    for num in range(1, 11):
        img_path = f"{king_en}{num}.jpg"
        img = cv2.imread(img_path)
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        
        faceCascade = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
        faces = faceCascade.detectMultiScale(gray, scaleFactor=1.12, minNeighbors=3, minSize=(50, 50))
        
        if len(faces) == 1:
            (x, y, w, h) = faces[0]
            crop_img = gray[y:y + h, x:x + w]
            resized_crop = cv2.resize(crop_img, (100, 100))  # 調整截圖大小，這裡使用 100x100
            
            img_train.append(resized_crop)
            labels.append(kings_en.index(king_en))
recon.train(img_train, np.array(labels))

img = cv2.imread("L13.jpg")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
faceCascade = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")

faces = faceCascade.detectMultiScale(
    gray,
    scaleFactor=1.12,
    minNeighbors=3,
    minSize=(50, 50)
)

print(faces)
print("發現", len(faces), "個人臉！！")

cut_list = []
cut_who = []

for (x, y, w, h) in faces:
    crop_img = img[y:y + h, x:x + w]
    cut_list.append(crop_img)

for cut_img in cut_list:
    predict_img = cv2.cvtColor(cut_img, cv2.COLOR_BGR2GRAY)
    label, confidence = recon.predict(predict_img)
    cut_who.append(kings[label])

for (x, y, w, h), who in zip(faces, cut_who):
    cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)
    cv2.putText(img, who, (x, y - 10), cv2.FONT_HERSHEY_TRIPLEX, 1, (0, 255, 0), 1)

cv2.imshow("findface", img)
cv2.waitKey()
```
